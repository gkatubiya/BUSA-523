```python
# Write a program for flipping a coin 10,000 times and store the results in a list
# After which, identify the number of streaks. (streak - a series of 5 or more
# heads or tails)

import random

streaks = 0
counter = 0
check_list = 0
current_list = 0

tossResults = []

for i in range(10000):
    toss = random.randint(0,1)
    if toss == 0:
        tossResults.append("heads")
    elif toss == 1:
        tossResults.append("tails")

for a in range(len(tossResults)):
    current_list = check_list
    check_list = tossResults[a]
    if current_list == check_list:
        counter += 1
    elif current_list != check_list:
        counter = 0
    if counter == 5:
        streaks += 1
        
streaks = str(streaks)
        
print("Number of streaks: " + streaks)
        
```
