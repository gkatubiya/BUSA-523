```python
#Problem 1
#Program which randomly selects a number between 1 and 20. 
# The user is given 5 chances to guess the number correctly to win. 

import random 

chancesTried = 0

print("Welcome to the Guessing Game. Please enter a number between 1 and 20")

number = random.randint(1,20)

while chancesTried < 5:
    user = input()
    user = int(user)
    chancesTried = chancesTried + 1
    
    if user != number: 
        print("Try again next time!")
        
    if user == number:
        break
        
if user == number:
    chancesTried = str(chancesTried)
    print("Congratulations, you are correct!")
if user != number:
          number=str(number)
          print("Game Over!")
```

```python
#Problem 2
# After each guess, print if the next guess should be lower or higher

import random 

chancesTried = 0

print("Welcome to the Guessing Game. Please enter a number between 1 and 20")

number = random.randint(1,20)

while chancesTried < 5:
    user = input()
    user = int(user)
    chancesTried = chancesTried + 1
    
    
    if user < number:
        print("Guess a higher value.")
    if user > number: 
        print ("Guess a lower value.")
    if user == number:
        break

if user == number:
    chancesTried = str(chancesTried)
    print("Congratulations, you are correct!")
        
if user != number:
    number=str(number)
    print("Better luck next time!")
    
    
```
